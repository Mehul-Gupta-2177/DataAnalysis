## The first data quality issues are the missing prices, because we do not know what that indicates. Does it mean that the Property is not available to sell, there is no valuation done on the property etc, Thus creating these confusions, which becomes an issue when finding the mean median and doign other analytics on the selling price say for a suburb. Another issue is that landsize of various houses are 0, thus making them false data.

## The second issue is that the variable distance is given but it is not explained to what the distance is relative to. It is from an absolute point from a suburb, or are all locations distanced from a particular area, Thus giving the person no scale as to how they should see the distance variable as and what is its use. # Figure out a way to replace missing values in bedroom columns and explain the way you used to replace the missing values and reason behind it. You can also select that no imputation possible but if this is this is the case then provide the reason behind it.

```python
df_bed_bath_cars_infer = df2.copy()
```

```python
df_bed_bath_cars_infer
```

```python
df_bed_bath_cars_infer['Bedroom'] = df_bed_bath_cars_infer['Bedroom'].fillna(df_bed_bath_cars_infer.groupby('Type')['Bedroom'].transform(lambda x: x.mode().iloc[0]))
```

### df_bed_bath_cars_infer # inferred number of bedrooms by housetype and took its mode because a house type needs a certain set of parameters thus inferring it this way will ensure that the rooms are in the average quartile for the house type. Same is done for cars and bathrooms

```python
df_bed_bath_cars_infer['Bathroom'] = df_bed_bath_cars_infer['Bathroom'].fillna(df_bed_bath_cars_infer.groupby('Type')['Bathroom'].transform(lambda x: x.mode().iloc[0]))
```

```python
df_bed_bath_cars_infer
```

```python
df_bed_bath_cars_infer['Car'] = df_bed_bath_cars_infer['Car'].fillna(df_bed_bath_cars_infer.groupby('Type')['Car'].transform(lambda x: x.mode().iloc[0]))
```

# Fill out the missing values in the councilarea column using suburb column

```python
f = lambda x: x.fillna(x.mode().iat[0]) if x.notna().any() else x
df_councilarea_infer['CouncilArea'] = df_councilarea_infer.groupby('Suburb')['CouncilArea'].apply(f)
```
